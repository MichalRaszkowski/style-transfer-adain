```python
import matplotlib.pyplot as plt
from datasets import load_dataset

# przykadowy dataset
content_dataset = load_dataset("CongWei1230/objectswap_coco_s0_100_samples", split="train", streaming=True)
style_dataset = load_dataset("newsletter/HiDream-I1-Artists", split="train", streaming=True)

# pobieramy kilka obrazw
for i, (content_sample, style_sample) in enumerate(zip(content_dataset, style_dataset)):
    content_image = content_sample["image"]  # PIL Image
    style_image = style_sample["image"]      # PIL Image

    # wywietlamy
    fig, axes = plt.subplots(1, 2, figsize=(8,4))
    axes[0].imshow(content_image)
    axes[0].set_title("Content")
    axes[0].axis("off")

    axes[1].imshow(style_image)
    axes[1].set_title("Style")
    axes[1].axis("off")

    plt.show()

    if i >= 3:  # tylko pierwsze 4 przykady
        break
```
